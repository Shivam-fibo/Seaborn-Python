# KDE PLOT -> Kernal Desity Estimation plot

```python

#  kdp plot produce the continuos density estimate
sns.kdeplot(data = tips, x ='total_bill')
#  y axis represnt the probabiltiy in the data
```

```python
#  kde plot using hiplot
sns.displot(data = tips , x='total_bill',kind = 'kde', hue = 'sex', fill = True)
```

#    RUGPLOT : Plot marginal distributuion by drawing ticks along the x and y axis


```python

sns.kdeplot(data = tips, x = 'total_bill')
sns.rugplot(data = tips, x = 'total_bill')
```


# bivarient histoplot -> it form a 2d plot in which the intersection of two plot give us information that there is a max number of frequecy of both points lies


```python
sns.histplot(data = tips, x = 'total_bill', y = 'tip')
#  for example the max number of bill is 15 where the max count of tips given of 2 doller
```
