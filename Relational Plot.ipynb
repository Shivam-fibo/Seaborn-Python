```python
# Relational Plot
#  . to see the static realtional between 2 or more variables
#  . Bivariante analysis
 
```

```python
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
#  scatter plot -> axis level function 
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip')
```

```python
#  relplot -> figure level -> square shape
sns.relplot(data = tips , x = 'total_bill', y= 'tip', kind = 'scatter')
plt.show()
```

```python
# property of seaborn -> hue ,style
sns.scatterplot(data = tips , x = 'total_bill', y = 'tip', hue = 'sex', style = 'time', size = 'size')
```

```python
gap = px.data.gapminder()
temp_df = gap[gap['country'] == 'India']
temp_df
x = temp_df['year']
x
y = temp_df['lifeExp']
y
plt.plot(x, y)
plt.xlabel('year')
plt.ylabel('lifeExp')
```

```python
#  lineplot -> axis level fucntion 
gap = px.data.gapminder()
temp_df = gap[gap['country'] == 'India']
temp_df
sns.lineplot(data = temp_df, x= 'year', y = 'lifeExp')
```

```python
#  figure level function 
sns.relplot(data = temp_df , x = 'year', y = 'lifeExp', kind = 'line')
```

```python
#  parameters
temp_df = gap[gap['country'].isin(['India', 'Germany','Brazil' ])]
temp_df
```

```python
sns.relplot(data = temp_df, x = 'year', y = 'lifeExp', hue = 'country',style = 'continent', kind = 'line', size = 'continent')
```

# Facet Plot

```python
sns.relplot(data = tips , x = 'total_bill', y= 'tip', kind = 'scatter', col = 'day',  col_wrap = 2)

```
