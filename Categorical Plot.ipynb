# Strip Plot :
## It take one categorical coulmn and one numerical coulmn and then plot in as scatter plot it is an axis level function

```python
tips
```

```python
sns.scatterplot(data = tips , x = 'total_bill', y = 'tip')
```

###  now we need to plot the scatter plot between total_bill and days

```python
sns.stripplot(data = tips, x = 'day' , y = 'total_bill', jitter = False)

```

###  seaborn add some noise to make clear picture of the plot if you want to remove it then :

```python
sns.catplot(data = tips, x = 'day' , y = 'total_bill', jitter = .1 , kind = 'strip')
```

##  Catplot : It is an figure level function 

## swan plot - > it help to visualize how the data is distributed on  a given range in this there is some internal algo running 

```python
sns.catplot(data = tips, x = 'day', y = 'total_bill', kind = 'swarm', hue = 'sex')
```

# Boxplot  : Way of displaying the distrubution of data based on five num summary minimum, median , maximum ,outliers and thier values

```python
# Box plotb
sns.boxplot(data = tips, x = 'day', y= 'total_bill')
```

```python
#  using catplot -> for figure level
sns.catplot(data = tips, x = 'day', y = 'total_bill' ,kind = 'box')
```

```python
sns.catplot(data = tips, y = 'total_bill', kind = 'box')
```
